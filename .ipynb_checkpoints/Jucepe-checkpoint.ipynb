## Importação de Bibliotecas

```python
import pandas as pd
import numpy as np
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
```

## Leitura do Arquivo - Abertura

```python
abertura = pd.read_csv('jucepe_abertura.csv', sep=";")

abertura
```

** 1º) O Total de Abertura da Jucepe

```python
abertura_total = abertura['TOTAL'].sum()
```

```python
tabela = pd.DataFrame({"Total" : [abertura_total]})

tabela["Total"] = tabela["Total"].map(lambda x: "R$ {:,.2f}".format(x))

tabela
```

** 2º) O Total de Abertura da Jucepe por Mês ?  

```python
abertura_total_mes = abertura.groupby(["MES"]).sum()['TOTAL'].rename("Total por Mês")
abertura_media_mes = abertura.groupby(["MES"]).mean()['TOTAL'].rename("Média por Mês")

analise_mes = pd.DataFrame({"Total por Mês" : abertura_total_mes, 
                            "Média por Mês" : abertura_media_mes})

analise_mes["Total por Mês"] = analise_mes["Total por Mês"].map(lambda x: "R$ {:,.2f}".format(x))
analise_mes["Média por Mês"] = analise_mes["Média por Mês"].map(lambda x: "R$ {:,.2f}".format(x))

analise_mes
```

** 3º) O Total de Abertura da Jucepe por Ano ?

```python
sns.set(style="darkgrid")
plt.figure(figsize=(10, 7))
sns.barplot(
    x="ANO", 
    y="TOTAL", 
    data=abertura, 
    estimator=sum, 
    ci=None, 
    color='#69b3a2');
```

** 4º) A média de Abertura da Jucepe por Ano ?

```python
abertura_media_ano = abertura.groupby(['ANO']).mean()['TOTAL'].rename("Média de Ano")

analise_ano = pd.DataFrame({"Média de Ano" : abertura_media_ano})

analise_ano["Média de Ano"] = analise_ano["Média de Ano"].map(lambda x: "R$ {:,.2f}".format(x))

analise_ano
```

## Leitura do Arquivo - Alteração

```python
alteracao = pd.read_csv('jucepe_alteracao.csv', sep=";")

alteracao
```

** 1º) Total de Alteração 

```python
total = alteracao['TOTAL'].sum()
total
```

** 2º) Total de Alteração por Mês 

```python
alteracao_por_mes = alteracao.groupby('MES')['TOTAL'].sum().reset_index()
```

```python
print(alteracao_por_mes)
```

```python
## Plotando Gráfico para Alteração do Mês
plt.figure(figsize=(10, 6))
plt.plot(alteracao_por_mes['MES'], alteracao_por_mes['TOTAL'], marker='o')
plt.title('Alteração de Impostos da Jucepe Mês')
plt.xlabel('Ano')
plt.ylabel('Total de Alterações')
plt.grid(True)
plt.show()
```

** 3º) Total de Alteração por Ano

```python
alteracao_por_ano = alteracao.groupby('ANO')['TOTAL'].sum().reset_index()
```

```python
print(alteracao_por_ano)
```

```python
## Plotando Gráfico para Alteração do Ano
plt.figure(figsize=(10, 6))
plt.plot(alteracao_por_ano['ANO'], alteracao_por_ano['TOTAL'], marker='o')
plt.title('Alteração de Impostos da Jucepe')
plt.xlabel('Ano')
plt.ylabel('Total de Alterações')
plt.grid(True)
plt.show()
```

** 4º) Média de Alteração da Jucepe

```python
media_mes_alteracao = alteracao['TOTAL'].mean()
```

```python
dicionario_media = pd.DataFrame({"Média do mês" : [media_mes_alteracao]})
dicionario_media.round(2)
```

** 5º) Alteração de Média por Mês 

```python
alteracao_media_mes = alteracao.groupby('MES')['TOTAL'].mean().reset_index()
```

```python
## Plotando Gráfico para Alteração do Mês
plt.figure(figsize=(10, 6))
plt.plot(alteracao_media_mes['MES'], alteracao_media_mes['TOTAL'], marker='o')
plt.title('Média de Alteração de Impostos da Jucepe no Mês')
plt.xlabel('Ano')
plt.ylabel('Média de Alterações')
plt.grid(True)
plt.show()
```

** 6º) Alteração de Média por Ano 

```python
alteracao_media_ano = alteracao.groupby('ANO')['TOTAL'].mean().reset_index()
```

```python
## Plotando Gráfico para Alteração do Mês
plt.figure(figsize=(10, 6))
plt.plot(alteracao_media_ano['ANO'], alteracao_media_ano['TOTAL'], marker='o')
plt.title('Média de Alteração de Impostos da Jucepe no Ano')
plt.xlabel('Ano')
plt.ylabel('Média de Alterações')
plt.grid(True)
plt.show()
```

## Leitura do Arquivo - Encerramento

```python
encerramento = pd.read_csv('jucepe_encerramento.csv', sep=';')
encerramento
```

** 1º) Total de Encerramento da Jucepe

```python
total_encerramento = encerramento['TOTAL'].sum()

dicionario_total = pd.DataFrame({"Total de Encerramento" : [total_encerramento]})

dicionario_total["Total de Encerramento"] = dicionario_total["Total de Encerramento"].map(lambda x: "R$ {:,.2f}".format(x))

dicionario_total
```

** 2º) Total de Encerramento por Ano

```python
encerramento_por_ano = encerramento.groupby('ANO')['TOTAL'].sum().reset_index()
```

```python
## Plotando Gráfico para Alteração do Mês
plt.figure(figsize=(10, 6))
plt.plot(encerramento_por_ano['ANO'], encerramento_por_ano['TOTAL'], marker='o')
plt.title('Total de Encerramento de Impostos da Jucepe no Ano')
plt.xlabel('Ano')
plt.ylabel('Total de Alterações')
plt.grid(True)
plt.show()
```

** 3º) Total de Encerramento por Mês

```python
sns.set(style="darkgrid")
plt.figure(figsize=(10, 7))
sns.barplot(
    x="MES", 
    y="TOTAL", 
    data=encerramento, 
    estimator=sum, 
    ci=None, 
    color='#69b3a2');
```

** 4º) Média de Encerramento por Mês

```python
encerramento_media_mes = encerramento.groupby(['MES']).mean()['TOTAL'].rename("Média do Mês")

analise_mes = pd.DataFrame({"Média do Mês" : encerramento_media_mes})

analise_mes["Média do Mês"] = analise_mes["Média do Mês"].map(lambda x: "R$ {:,.2f}".format(x))

analise_mes
```

** 5º) Média de Encerramento por Ano

```python
encerramento_media_ano = encerramento.groupby(['ANO']).mean()['TOTAL'].rename("Média do Ano")

analise_ano = pd.DataFrame({"Média do Ano" : encerramento_media_ano})

analise_ano["Média do Ano"] = analise_ano["Média do Ano"].map(lambda x: "R$ {:,.2f}".format(x))

analise_ano
```
